```python
"""
Extract spectra used in Figure 5
"""

spectra_bnorm = []
spectra_nonorm = []
for region in ['MIL2', 'CCTrees', 'Puako']:
    spectra = figz.extract(f'{region}_spectra.shp', f'{region}_refl.tif', brightness_norm=True)
    spectra_bnorm.append(spectra)
    spectra = figz.extract(f'{region}_spectra.shp', f'{region}_refl.tif', brightness_norm=False)
    spectra_nonorm.append(spectra)
```

```python
%%time
"""
Create Figure 5
"""

figz.shapley_plot(spectra_bnorm, spectra_nonorm, ptrue=0.01)
```
